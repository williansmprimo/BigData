```python
from pyspark.sql import SparkSession
spark = SparkSession.builder.appName('kmeans').getOrCreate()
```

```python
data = spark.read.csv('customer_segmentation.csv', header=True, inferSchema=True)
```

```python
data.printSchema()
```

```python
columns = ['Income',
 'Kidhome',
 'Teenhome',
 'MntWines',
 'MntFruits',
 'MntMeatProducts',
 'MntFishProducts',
 'MntSweetProducts',
 'MntGoldProds',
 'NumDealsPurchases',
 'NumCatalogPurchases',
 'NumStorePurchases',
 'NumWebVisitsMonth',
 'AcceptedCmp3',
 'Complain',
 'Z_CostContact']
```

```python
from pyspark.ml.clustering import KMeans
from pyspark.ml.feature import VectorAssembler
from pyspark.ml.feature import StandardScaler
assembler = VectorAssembler(inputCols=columns, outputCol='features_')
transformed_data = assembler.transform(data).na.fill(0).select('features_')
scaler_ = StandardScaler(inputCol='features_', outputCol='features')
scaler = scaler_.fit(transformed_data)
transformed_data = scaler.transform(transformed_data)
transformed_data = transformed_data.select('features')
```

```python
kmeans = KMeans(featuresCol='features',k=2).setSeed(1)
```

```python
model = kmeans.fit(transformed_data)
centers = model.clusterCenters()
```

```python
len(centers), len(centers[0])
```

```python
transformed_data = model.transform(data)
```
