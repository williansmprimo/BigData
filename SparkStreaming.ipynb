```python
from pyspark import SparkContext
from pyspark.streaming import StreamingContext
```

```python
sparkContext = SparkContext('local[2]', 'StreamingExample1')
streamingContext = StreamingContext(sparkContext, 1)
```

```python
socket = streamingContext.socketTextStream('localhost', 9999)
```

```python
myWords = socket.flatMap(lambda line: line.split(' '))
```

```python
pairs = myWords.map(lambda word: (word, 1))
```

```python
words_count = pairs.reduceByKey(lambda n1, n2: n1 + n2)
```

```python
words_count.pprint()
```

```python
streamingContext.start()
```

```python
streamingContext.stop()
```
